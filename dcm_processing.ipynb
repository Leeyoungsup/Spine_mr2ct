```python
import pydicom as dcm
import numpy as np
import pandas as pd
import os
from tqdm import tqdm
from glob import glob
```

```python
ct_df= pd.DataFrame(columns=['ID','thickness','pixelspacing','depth'])
mri_df= pd.DataFrame(columns=['ID','thickness','pixelspacing','depth'])


ct_folder_list=glob('../../data/250513 C CT/**')
mri_folder_list=glob('../../data/250513 C MR/**')
```

```python
for i in tqdm(range(len(ct_folder_list))):
    ct_folder=ct_folder_list[i]
    ct_files=glob(ct_folder+'/*.dcm')
    depth=len(ct_files)
    dcm_file=dcm.dcmread(ct_files[0])
    thickness=dcm_file.SliceThickness
    pixel_spacing=dcm_file.PixelSpacing[0]
    id_s=os.path.basename(ct_folder)
    ct_df.loc[i]=[id_s,thickness,pixel_spacing,depth]

ct_df.to_csv('../../data/ct_df.csv',index=False,encoding='utf-8-sig')

for i in tqdm(range(len(mri_folder_list))):
    mri_folder=mri_folder_list[i]
    mri_files=glob(mri_folder+'/*.dcm')
    depth=len(mri_files)
    dcm_file=dcm.dcmread(mri_files[0])
    thickness=dcm_file.SliceThickness
    pixel_spacing=dcm_file.PixelSpacing[0]
    id_s=os.path.basename(mri_folder)
    mri_df.loc[i]=[id_s,thickness,pixel_spacing,depth]
mri_df.to_csv('../../data/mri_df.csv',index=False,encoding='utf-8-sig')
```

```python
filter_ct_df=ct_df[ct_df['thickness']==2.]
filter_ct_df=filter_ct_df[filter_ct_df['pixelspacing']<1]
filter_ct_df.to_csv('../../data/filter_ct_df.csv',index=False,encoding='utf-8-sig')
filter_mri_df=mri_df[mri_df['thickness']==3.]
filter_mri_df=filter_mri_df[filter_mri_df['pixelspacing']<1]
filter_mri_df.to_csv('../../data/filter_mri_df.csv',index=False,encoding='utf-8-sig')
```
